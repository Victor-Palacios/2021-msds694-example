```python
from pyspark.sql import SparkSession
from pysparkling import *

ss = SparkSession.builder.getOrCreate()
hc = H2OContext.getOrCreate()
```

## Create DataFrame from parquet file and convert to H2OFrame

### Parquet files with nested structures are not supported.

```python
import h2o
#adult_train_df = h2o.import_file("../Data/spark-warehouse/adulttrain") 
```

```python
adult_train_df = ss.read.parquet("../Data/spark-warehouse/adulttrain")
```

```python
adult_train_h2o = hc.asH2OFrame(adult_train_df, "adult_train")
```

```python
adult_train_h2o
```

```python
adult_train_h2o["label"] = adult_train_h2o["label"].asfactor()
#adult_valid_h2o["label"] = adult_valid_h2o["label"].asfactor()
```

```python
adult_train_h2o.types
```

```python
adult_train_h2o
```

```python
adult_valid_df = ss.read.parquet("../Data/spark-warehouse/adultvalid")
adult_valid_h2o = hc.asH2OFrame(adult_valid_df, "adult_valid")

adult_valid_h2o["label"] = adult_valid_h2o["label"].asfactor()
adult_valid_h2o.summary()
```

## Apply ML Algorithms 
Algorithm Parameter Details : http://docs.h2o.ai/h2o/latest-stable/h2o-py/docs/_modules/index.html

### Define Predictors

```python
predictors = adult_valid_h2o.names[:]
response = "label"
predictors.remove(response)
```

```python
predictors
```

### Gradient Boosting
http://docs.h2o.ai/h2o/latest-stable/h2o-py/docs/_modules/h2o/estimators/gbm.html

http://docs.h2o.ai/h2o/latest-stable/h2o-py/docs/modeling.html#h2o.estimators.gbm.H2OGradientBoostingEstimator


```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator
model_gbm = H2OGradientBoostingEstimator(ntrees=50,
                                         max_depth=6,
                                         learn_rate=0.1,
                                         distribution="bernoulli")

model_gbm.train(x=predictors,
                y=response,
                training_frame=adult_train_h2o,
                validation_frame=adult_valid_h2o)
```

```python
model_gbm
```

```python
print(model_gbm.confusion_matrix(train=True))
print(model_gbm.confusion_matrix(valid=True))
```

```python
print(model_gbm.auc(train=True))
print(model_gbm.auc(valid=True))
```

### Deep Learning
http://docs.h2o.ai/h2o/latest-stable/h2o-py/docs/_modules/h2o/estimators/deeplearning.html

http://docs.h2o.ai/h2o/latest-stable/h2o-py/docs/modeling.html#h2o.estimators.deeplearning.H2ODeepLearningEstimator

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
model_dl = H2ODeepLearningEstimator(variable_importances=True,
                                    loss="Automatic")

model_dl.train(x=predictors,
               y="label",
               training_frame=adult_train_h2o,
               validation_frame=adult_valid_h2o)
```

```python
model_dl
```

```python
print(model_dl.confusion_matrix(train=True))
print(model_dl.confusion_matrix(valid=True))
print(model_dl.auc(train=True))
print(model_dl.auc(valid=True))
```

### XG Boosting
http://docs.h2o.ai/sparkling-water/2.4/latest-stable/doc/tutorials/sw_xgboost.html

http://docs.h2o.ai/h2o/latest-stable/h2o-py/docs/_modules/h2o/estimators/xgboost.html

http://docs.h2o.ai/h2o/latest-stable/h2o-py/docs/modeling.html#h2o.estimators.xgboost.H2OXGBoostEstimator

```python
from h2o.estimators.xgboost import H2OXGBoostEstimator
model_xg = H2OXGBoostEstimator()
model_xg.train(x=predictors,
               y="label",
               training_frame=adult_train_h2o,
               validation_frame=adult_valid_h2o)
```

```python
model_xg
```

```python
print(model_xg.confusion_matrix(train=True))
print(model_xg.confusion_matrix(valid=True))
print(model_xg.auc(train=True))
print(model_xg.auc(valid=True))
```
